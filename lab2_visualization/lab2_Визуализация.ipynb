## Задание 1.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Загрузка данных
url = 'https://raw.githubusercontent.com/chponya/python_labs/refs/heads/main/lab2_visualization/company_sales_data.csv'
df = pd.read_csv(url)
```

```python
# Построение линейного графика
plt.figure(figsize=(10, 5))
plt.plot(df['month_number'], df['total_profit'], color='blue', linestyle='-')

# Настройка графика
plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')
plt.xticks(df['month_number'])  # чтобы X были от 1 до 12

# Показать график
plt.tight_layout()
plt.show()
```

## Задание 2.

```python
# Построение графика
plt.figure(figsize=(10, 5))
plt.plot(
    df['month_number'],
    df['total_units'],
    color='red',
    marker='o',
    linestyle='--',
    linewidth=3,
    label='Profit data of last year'
)

# Настройки графика
plt.title('Company Sales data of last year')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')

plt.legend(loc='lower right')
plt.xticks(df['month_number'])

# Показать график
plt.tight_layout()
plt.show()
```

## Задание 3.

Часть 1: Один график со всеми линиями

```python
month = df['month_number']

plt.plot(month, df['facecream'], label='Face cream Sales Data', marker='o')
plt.plot(month, df['facewash'], label='Face Wash Sales Data', marker='o')
plt.plot(month, df['toothpaste'], label='ToothPaste Sales Data', marker='o')
plt.plot(month, df['bathingsoap'], label='BathingSoap Sales Data', marker='o')
plt.plot(month, df['shampoo'], label='Shampoo Sales Data', marker='o')
plt.plot(month, df['moisturizer'], label='Moisturizer Sales Data', marker='o')

plt.title('Sales data')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Часть 2: Отдельные графики для каждого продукта

```python
fig = plt.figure(figsize=(12, 10))

def product_plots(i):
    plt.subplot(3, 2, i+1)
    plt.plot(month, df[units[i]], color=colors[i], marker='o')
    plt.title(f'Sales data of {names[i]}')
    plt.xlabel('Month Number')
    if i%2==0:
        plt.ylabel('Sales units')

units = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
colors = ['blue', 'red', 'green', 'black', 'magenta', 'brown']
names = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

for i in range(6):
    product_plots(i)

plt.tight_layout()
plt.show()
```

## Задание 5.

```python
month = df['month_number']
face_cream = df['facecream']
face_wash = df['facewash']

bar_width = 0.3

# Позиции для столбцов
x = month
x2 = [i + bar_width for i in x]

# Построение графика
plt.bar(x, face_cream, width=bar_width, label='Face Cream sales data', color='blue')
plt.bar(x2, face_wash, width=bar_width, label='Face Wash sales data', color='orange')

# Настройка графика
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Facewash and facecream sales data')
plt.legend()
plt.grid(True, linestyle='--')
plt.xticks(df['month_number'])  # чтобы X были от 1 до 12

# Показываем график
plt.show()
```

## Задание 7.

stacked plot (слоеная диаграмма)

```python
# Месяцы
month = df['month_number']

# Продукты
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

# Построение stacked plot
plt.figure(figsize=(10, 6))
plt.stackplot(
    month,
    facecream,
    facewash,
    toothpaste,
    bathingsoap,
    shampoo,
    moisturizer,
    labels=[
        'face Cream', 'Face wash', 'Tooth paste',
        'Bathing soap', 'Shampoo', 'Moisturizer'
    ],
    colors=['magenta', 'cyan', 'red', 'black', 'green', 'yellow']
)

plt.legend(loc='upper left')
plt.title('All product sales data using stack plot')
plt.xlabel('Month Number')
plt.ylabel('Sales units in Number')
plt.tight_layout()
plt.show()
```

## Задание 8.

Заготовка для расположения графиков

```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Создаем фигуру
fig = plt.figure(figsize=(10, 8))
gs = gridspec.GridSpec(2, 3, height_ratios=[2, 1])  # верхний ряд выше

# Верхний график (на всю ширину)
ax1 = fig.add_subplot(gs[0, :])
ax1.set_title('Top plot')

# Нижние три графика
ax2 = fig.add_subplot(gs[1, 0])
ax2.set_title('Bottom left')

ax3 = fig.add_subplot(gs[1, 1])
ax3.set_title('Bottom center')

ax4 = fig.add_subplot(gs[1, 2])
ax4.set_title('Bottom right')

# Добавим рамки (для визуализации заготовки)
for ax in [ax1, ax2, ax3, ax4]:
    ax.plot([], [])  # пустые графики
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
